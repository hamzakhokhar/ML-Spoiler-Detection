```python
import pandas as pd 
import numpy as np
from IPython.display import display
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.stem.porter import PorterStemmer
from nltk.stem.snowball import SnowballStemmer
import spacy
nlp = spacy.load("en_core_web_lg")
```

```python
df = pd.read_json('/Users/hamza/ml-project/IMDB_reviews.json', lines = True)
md = pd.read_json('/Users/hamza/ml-project/IMDB_movie_details.json', lines = True)
```

```python
csynopsis = pd.read_csv('Cutdown Synopsis.csv')
```

```python
syn = list(csynopsis['movie_id'])
```

```python
combined = pd.merge(df, md, on = 'movie_id')
```

```python
new_df = combined[combined.movie_id.isin(syn)]
```

```python
new_df = new_df.drop(['review_date', 'user_id', 'is_spoiler',
       'rating_x', 'review_summary', 'plot_summary', 'duration', 'genre',
       'rating_y', 'release_date'], axis = 1)
```

```python
review_list = list(new_df['review_text'])
review_list = review_list[0:50000]
synopsis_list = list(new_df['plot_synopsis'])
synopsis_list = synopsis_list[0:50000]
```

```python
import time
start = time.time()
scores = []
for i in range(0, len(synopsis_list)):
    doc1 = nlp(review_list[i])
    doc2 = nlp(synopsis_list[i])
    scores.append(doc2.similarity(doc1))
elapsed_time_fl = (time.time() - start) 
print(elapsed_time_fl)
print(scores)   
```

```python
df = pd.DataFrame(data = scores)
df.to_csv('0-50000.csv', index = False)
```
